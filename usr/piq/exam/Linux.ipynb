### ````chmod```` et ````chown````

Les deux commandes permettent de grer les permissions attribues au fichiers ou dossiers. Plus prcisment, ````chown```` change le propritaire d'un fichier ou d'un rpertoire. Pour ce faire, il suffit d'entrer la commande suivante: ````chown propritaire:groupe le_fichier_cible````.
La commande ````chmod```` est plus complexe, en effet, elle fournit deux types de syntaxe pour la dfinission des permissions.
La premire, la plus logique, change les permissions de manire relatives. C'est  dire que soit vous ajoutez des permissions, soit vous en retirez. La syntaxe et la suivante ````chmod QUI+/-permission nomdufichier````, o il faut remplacer "qui" par:
* "a" si vous voulez modifier les permissions pour tout le monde
* "u" si vous voulez modifier les permissions pour le propritaire
* "g" si vous voulez modifier les permissions pour le groupe auxquel appartient le fichier
* "o" si vous voulez modifier les permissions pour les autres utilisateurs

Ensuite, vient soit le signe plus (+) pour ajouter des permissions, soit le signe moins(-) pour en retirer.
Puis les permissions  ajouter ou retirer: 
* "r" (read), la permission de lecture
* "w" (write), la permission d'criture
* "x", la permission d'xecuter

Par exemple, pour attribuer la permissions d'xecution  tout les utilisateurs, la commande requise est ````chmod a+x nomdufichier````.

Passons maintenant  la deuxime mthode, qui dfini compltement les permissions. Cette mthode s'appuie sur le nombre  trois chifffres voqu dans la partie traitant de la commande ````mkdir````. Le premier chiffre gre les permissions du propritaire, le deuxime celles du groupe, et le troisime les permissions des autres uilisateurs. Les chiffres quant  eux reprsentent les ppermissions selon le tableau suivant :

|       Permissions                                    | Valeur littral| Valeur numrique |
|:---:|:---:|:---:|
|                         Aucune Permission            | ---            | 0                |
| xecution seulement                                  | --x            | 1                |
| criture seulement                                   | -w-            | 2                |
| xecution et criture                                | -wx            | 3                |
| Lecture seulement                                    | r--            | 4                |
| Lecture et xecution                                 | r-x            | 5                |
| Lecture et criture                                  | rw-            | 6                |
| Toutes les permissions (lecture, criture, xecution)| rwx            | 7                |

Ainsi la commande ````chmod 740 nomdufichier```` donne tout les droits au propritaire, le droit de lecture uniquement aux membres du groupe, et aucun droit sur ce fihcier aux autres utilisateurs.

Par dfaut, (en tout cas sur le Raspebrry Pi), le temps d'attente est de 1 minute.
Enfin, si vous voulez annuler une extinction planifie, la commande ````shutdown -c```` s'en chargera.
