## Annotations

https://matplotlib.org/tutorials/text/annotations.html

The `text()` command can be used to add text in an arbitrary location, and the `xlabel()`, `ylabel()` and `title()` are used to add text in the indicated locations.

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
r = np.arange(0,1,0.001)
theta = 2 * 2*np.pi * r
line, = ax.plot(theta, r, color='#ee8d18', lw=3)

ind = 700
thisr, thistheta = r[ind], theta[ind]
ax.plot([thistheta], [thisr], 'o')
ax.annotate('a polar annotation',
            xy=(thistheta, thisr),  # theta, radius
            xytext=(0.05, 0.05),    # fraction, fraction
            textcoords='figure fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            )
plt.show()
```

## text()

```python
kwargs = dict(fontsize=14, bbox=dict(boxstyle="round", fc="pink", pad=0.8))

fig, ax = plt.subplots(figsize=(10, 5))
ax.set_aspect(1)

plt.axis([0, 6, 0, 2])
ax.set_aspect(1)
for x in range(6):
    for y in range(2):
        plt.text(x, y, 'cell {}{}'.format(x, y), **kwargs)
        
plt.grid(False)
plt.show()
```

annotate with `bbox`

```python
bbox1 = dict(boxstyle="rarrow,pad=0.3", fc="cyan", ec="b", lw=2)
plt.text(0, 0, "Direction", ha="center", va="center", size=15, bbox=bbox1)

bb2 = dict(boxstyle="rarrow,pad=0.3", fc="cyan", ec="b", lw=2)
plt.text(1, 0, "Direction", ha="center", va="center", size=15, bbox=bb2)

plt.show()
```

```python

labels = ['C1', 'C2', 'C3', 'C4']
kwargs = dict(va="center", ha="left", textcoords='data', size=16,
              bbox=dict(boxstyle="round", fc="pink"), 
              arrowprops=dict(arrowstyle="->"))
fig, ax = plt.subplots(figsize=(12, 2))

    
an1 = ax.annotate(labels[0], xy=(0.2, 0.2), **kwargs)
an2 = ax.annotate(labels[1], xy=(1, 0.5), xycoords=an1, xytext=(0.3, 0.2), **kwargs)
an2 = ax.annotate(labels[2], xy=(1, 0.5), xycoords=an2, xytext=(0.4, 0.2), **kwargs)
an3 = ax.annotate(labels[3], xy=(0.8, 1), xycoords=an2, xytext=(0.4, 0.8), **kwargs)



an3 = ax.annotate(labels[2], xy=(0.5, 1), xycoords=an2,
                  xytext=(0, 50), textcoords="offset points",
                  va="center", ha="center",
                  bbox=dict(boxstyle="square", fc="lime"),
                  arrowprops=dict(arrowstyle="->"))
plt.show()
```
